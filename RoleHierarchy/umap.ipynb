```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from minedgerolemining.readup import readup,readup_and_usermap_permmap
```

```python
upfilename = '/home/puneet/Projects/minedgerolemining/minedgerolemining/inputsup/UP-hc'
```

```python
up, usermap, permmap = readup_and_usermap_permmap(upfilename)
inv_usermap = {v: k for k, v in usermap.items()}
inv_permmap = {v: k for k, v in permmap.items()}
```

```python
inv_permmap
```

```python
up_df = pd.DataFrame()
rows = list()
for u in up:
    row = {'user': inv_usermap[u]}
    for p in inv_permmap:
        row[inv_permmap[p]] = 0
    for p in up[u]:
        row[inv_permmap[p]] = 1
    print(row)
    rows.append(row)
# print(rows)
```

```python
up_df = pd.DataFrame(rows)
```

```python
up_df
```

```python
sns.pairplot(up_df, hue='user');
```

```python
import umap

reducer = umap.UMAP()
scaled_up_data = StandardScaler().fit_transform(up_df)

```

```python
embedding = reducer.fit_transform(scaled_up_data)
embedding.shape
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in up_df.species.map({"Adelie":0, "Chinstrap":1, "Gentoo":2})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Penguin dataset', fontsize=24);
```
